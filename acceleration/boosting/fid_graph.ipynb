```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fidg = np.load("fid_list_data/data_gan_fid.npy")
```

```python
fidgp = np.load("fid_list_data/data_gan_fid_perc.npy")
```

```python
fidgl = np.load("fid_list_data/data_gan_fid_lossf.npy")
```

```python
fidgr = np.load("fid_list_data/data_gan_fid_random.npy")
```

```python
fidv2 = np.load("fid_list_data/data_gan_fid_lossfv2.npy")
```

```python
fidv3 = np.load("fid_list_data/data_gan_fid_lossfv3.npy")
```

```python
fidv4_03 = np.load("fid_list_data/data_gan_fid_lossfv4_exact.npy")
```

```python
fidv4_2 = np.load("fid_list_data/data_gan_fid_lossfv4_exact2.npy")
```

```python
fidv4_1 = np.load("fid_list_data/data_gan_fid_lossfv4_exact1.npy")
```

```python
fidv4_01 = np.load("fid_list_data/data_gan_fid_lossfv4_exact01.npy")
```

```python
def plot_array_sampled(arr, N, label=None):
    x = list(range(0, 10 * (N + 1), 10))
    y = arr[0:N + 1]
    
    plt.plot(x, y, label=label)
    plt.xlabel("No. of images augumented")
    plt.ylabel("FID")
    plt.title("FID vs Images added")
    plt.grid(True)
```

```python
N = 100
plt.figure(figsize=(12, 8))
plot_array_sampled(fidg, N, label="GAN")
plot_array_sampled(fidgp, N, label="GAN perceptual")
plot_array_sampled(fidgl, N, label="GAN_fid")
plot_array_sampled(fidgr, N, label="Random")
plot_array_sampled(fidv2, N, label="fid mean")
plot_array_sampled(fidv3, N, label="fid mean2")
plot_array_sampled(fidv4_03, N, label="fid exact l=0.3")
plot_array_sampled(fidv4_2, N, label="fid exact l=2")
plot_array_sampled(fidv4_1, N, label="fid exact l=1")
plot_array_sampled(fidv4_01, N, label="fid exact l=0.1")
plt.legend()
plt.show()
```
